```python
import pandas as pd

# Leer el archivo country_vaccinations.csv
data = pd.read_csv("country_vaccinations.csv")

# Convertir columnas con fechas al tipo datetime64
data["date"] = pd.to_datetime(data["date"])

# Mostrar la estructura y tipos de datos de cada columna despues de la modificacion
print("Estructura y tipos de datos: \n")
print(data.info(), "\n")
print("************************************************************************************")

# Determinar la cantidad de vacunas aplicadas de cada compañía
print("Vacunas aplicadas por compañia: \n")
vacunas_por_compañia = data.groupby("vaccines").size().reset_index(name="cantidad_aplicada")
print(vacunas_por_compañia, "\n")

# Obtener la cantidad total de vacunas aplicadas en todo el mundo
cantidad_total_vacunas = data["total_vaccinations"].sum()
print("************************************************************************************")
print("La cantidad total de vacunas aplicadas en todo el mundo es: ", cantidad_total_vacunas)
print("************************************************************************************")

# Calcular el promedio de vacunas aplicadas por país
print("Promedio de vacunas aplicadas por país: \n")
promedio_vacunas_por_pais = data.groupby("country")["total_vaccinations"].mean().reset_index().round(0)
print(promedio_vacunas_por_pais)
print("************************************************************************************")

# Determinar la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo.
print("Vacunas aplicadas el dia 29/01/21 en todo el mundo: \n")

vacunas_aplicadas_29_01_21 = data[(data["date"] == "2021-01-29")]["total_vaccinations"].sum()
print(vacunas_aplicadas_29_01_21)
print("************************************************************************************")

# Creación de un dataframe nuevo con la columna de diferencias
print("Creación de un dataframe nuevo con la columna de diferencias: \n")

# Crear el nuevo dataframe con una columna derivada 'diferencias'
conDiferencias = data.copy()  # Copiar el dataframe original

# Calcular la diferencia entre las columnas 'daily_vaccinations' y 'daily_vaccinations_raw' y almacenarla en la nueva columna 'diferencias'
conDiferencias["diferencias"] = data["daily_vaccinations"] - data["daily_vaccinations_raw"]

print(conDiferencias)
print("************************************************************************************")

# Obtener el periodo de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua.
periodo_de_tiempo = data["date"].max() - data["date"].min()
if periodo_de_tiempo.days == 1:
    print("El período de tiempo entre el registro con fecha más reciente y el registro con fecha más antigüa es de: 1 día")
else:
    print(f"El período de tiempo entre el registro con fecha más reciente y el registro con fecha más antigüa es de: {periodo_de_tiempo.days} días")
print("************************************************************************************")

print("Creación de un dataframe nuevo con la cantidad de vacunas utilizadas cada día: \n")
# Cargar el archivo country_vaccinations.csv en un dataframe
df = pd.read_csv("country_vaccinations.csv")

# Crear el dataframe con la columna derivada
conCantidad = df.copy()

# Crear un nuevo dataframe con los datos originales
conCantidad["canVac"] = conCantidad["vaccines"].str.split(",").apply(lambda x: len(x))

# Mostrar el nuevo dataframe con los datos originales y la columna derivada
print(conCantidad[["vaccines", "canVac"]])
print("************************************************************************************")

print("Creación de un dataframe nuevo denominado antes20 con todos los registros que se hayan realizado antes del 20 de diciembre de 2020: \n")
# Cargar el archivo CSV en un DataFrame
df = pd.read_csv("country_vaccinations.csv")

# Crear el DataFrame antes20 con registros antes del 20 de diciembre de 2020
antes20 = df[df["date"] < "2020-12-20"]

# Verificar el resultado
print(antes20)
print("************************************************************************************")

print("Creación de un dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna Pfizer: \n")
# Cargar el archivo CSV en un DataFrame
df = pd.read_csv("country_vaccinations.csv")

# Crear el DataFrame pfizer con los registros de la vacuna Pfizer
pfizer = df[df["vaccines"].str.contains("Pfizer", case=False, na=False)]

# Verificar el resultado
print(pfizer)
print("************************************************************************************")

print("Almacenar dataframes generados (conDiferencias, conCantidad y pfizer) \n")
# Almacenar dataframes en Excel
with pd.ExcelWriter('resultadosReto.xlsx') as writer:
    conDiferencias.to_excel(writer, sheet_name='conDiferencias', index=False)
    conCantidad.to_excel(writer, sheet_name='conCantidad', index=False)
    antes20.to_excel(writer, sheet_name='antes20', index=False)
    pfizer.to_excel(writer, sheet_name='pfizer', index=False)
```
